#### Rotten Tomatoes ([Tomatometer](https://www.rottentomatoes.com/about))
수백명의 영화나 드라마 평론가들의 의견들을 바탕으로 수백만 팬들에게 믿을만한 추천하는 정도를 표시하는 방법.
즉, Tomatometer Score는 전문적인 평론가들이 영화나 TV show에 긍정적인 평가의 수를 전체 평가의 수에 대해 보여주는 백분율을 의미한다. 영화나 TV show에 대한 리뷰 중 60%가 긍정적인 리뷰라면 Rotten Tomatoes에서는 빨간 토마토를 보여주면서 Fresh status라고 표시되지만, 60% 미만이라면, green splat를 보여주면서 Rotten status라고 표시된다. 

Rotten Tomatoes에서는 Audience Score도 보여준다. Tomatometer에서 토마토로 표시되는 평론가들의 점수와 비슷하게 Audience Score는 팝콘으로 표시되며 영화나 TV show에 대한 긍정적인 평가의 백분율을 의미한다. Rotten Tomatoes 사용자들 중 60%가 영화나 TV show에 별 3.5개 이상을 주었다면, 팝콘이 가득 채워져 있는 이미지를 보여주면서 Fresh status라고 표시되지만, 비슷하게 60% 가 별 3.5개 미만을 주었다면 엎어져 있는 팝콘을 보여주며 Rotten status라고 표시된다.

(뒤에 사용할 데이터에는 status를 볼 수 없고 백분율로 정리된 점수만 사용할 예정이다.)

Rotten Tomatoes (Tomatometer)는 사람들이 영화가 어떤지 TV show가 어떤지 점수나 이미지를 보고 직접 시청하기 전에 어느정도 판단이 가능하고 이는 사람들에게 영화나 드라마 TV show를 추천하는데 믿을 만한 정보이자 근거가 된다.

```python
# Extract three columns from df_mcu and Create new data
df_Score = df_mcu[['movie_title','tomato_meter','audience_score']]

# Create DataFrame
df_Score_1 = pd.DataFrame(df_Score)

# Create a Dictionary
dict_df_Score = pd.DataFrame(df_Score).set_index('movie_title').to_dict(orient='index')



# Define function for Finding Outlier using Dictionary
def FindingOutlier(col):
    for i in df_Score['movie_title']:
        if dict_df_Score[i][col] == min(df_Score[col]):
            return f"Movie title of Outlier: {i}"

# Outlier 1(tomato_meter가 가장 낮은 영화 제목) 찾기
print(FindingOutlier('tomato_meter'))
        
# Outlier 2(audience_score가 가장 낮은 영화 제목) 찾기
print(FindingOutlier('audience_score'))
```

### Proposition 3. 
### Domestic Box Office가 높으면 높을수록 Worldwide Box Office도 높을 것이다.

```python
# 두 개의 columns만 따로 추출하여 새로운 표 만들기
df_mcu_p3 = df_mcu[['domestic_box_office','worldwide_box_office']]
df_mcu_p3
```

```python
# 문자형에서 실수형으로 데이터값 변환하기
df_mcu_p3['domestic_box_office'] = df_mcu_p3['domestic_box_office'].str.replace(',','').astype(float)
df_mcu_p3['worldwide_box_office'] = df_mcu_p3['worldwide_box_office'].str.replace(',','').astype(float)

df_mcu_p3
```

.astype() method를 이용하여 문자형이었던 box_office의 숫자들을 계산할 수 있는 숫자들의 형태로 바꾸어주었는데, 여기서 다른 경우들과는 다르게 float형으로 변형하였다. 
float형으로 변형한 이유는 숫자 데이터가 너무 커서 정수형으로 변환할 수 없어 Overflow Error를 발생시킬 수 있기 때문이다. 실수형은 정수형보다 큰 데이터를 다룰 수 있지만, 데이터의 정밀도가 떨어질 수 있음에 유의한다.

```python
# Create DataFrame
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df_boxoffice_p3 = pd.DataFrame(df_mcu_p3)

# Draw Scatterplot
df_boxoffice_p3.plot(kind = 'scatter', x = 'domestic_box_office', y = 'worldwide_box_office')

# Draw Linear Regression
b0, b1 = np.polyfit(df_boxoffice_p3['domestic_box_office'], df_boxoffice_p3['worldwide_box_office'], 1)
plt.plot(df_boxoffice_p3['domestic_box_office'], b1+b0*df_boxoffice_p3['domestic_box_office'])

plt.show()
```

domestic_box_office와 worldwide_box_office 사이의 관계를 표현하는 Scatterplot에서 데이터들을 시각적으로 이어보면 뚜렷한 하나의 선이 생긴다. domestic_box_office와 worldwide_box_office는 비례적인 관계임을 보여준다. 즉, 국내(미국기준)에서의 박스오피스가 높으면 높을수록 해외에서의 박스오피스도 높다는 것을 의미한다.

Linear Regression을 데이터들 사이의 차이가 최소화되는 직선으로 domestic_box_office와 worldwide_box_office 사이의 관계를 시각적으로 보여주었다. 위에서 얘기한 하나의 선은 Linear Regression을 의미하고, 이를 표시하기 위해 조사를 통해 Linear Regression을 구현하였다.

**Conclusion of Proposition 3:**

Domestic Box Office가 높으면 높을수록 Worldwide Box Office도 높았다.  (영화의 수익이 국내(미국기준)에서 좋으면 좋을수록 해외(전세계)에서도 수익이 좋았다.)
